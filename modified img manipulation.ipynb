```python
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt

my_image1=Image.open('C:/images/dog.jpg')

plt.figure(figsize=(10,10))
plt.subplot(2,2,1);
plt.title("Original image")
plt.imshow(my_image1);
plt.axis('off')

sharp1=my_image1.filter(ImageFilter.SHARPEN);
plt.subplot(2,2,2);
plt.title("Sharpend image")
plt.imshow(sharp1);
plt.axis('off')


img=Image.open('C:/images/travel.jpg')

plt.subplot(2,2,3);
plt.title("Original image")
plt.imshow(img);
plt.axis('off')

imgs=img.filter(ImageFilter.SHARPEN);
plt.subplot(2,2,4);
plt.title("Sharpend image")
plt.imshow(imgs);
plt.axis('off')


```
